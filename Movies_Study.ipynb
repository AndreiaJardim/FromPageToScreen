1. Most profitable movies of all time that are original screenplay

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# Set up Chrome driver options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode

# Set path to chromedriver executable
webdriver_path = 'path/to/chromedriver'

# Start the Selenium Chrome driver
driver = webdriver.Chrome(service=Service(executable_path=webdriver_path), options=chrome_options)

# Website URL
url = 'https://www.the-numbers.com/box-office-records/worldwide/all-movies/sources/original-screenplay'

# Open the webpage
driver.get(url)

# Find the table element using XPath
table = driver.find_element(By.XPATH, '//*[@id="page_filling_chart"]/center/table')

# Extract the rows from the table
rows = table.find_elements(By.TAG_NAME, 'tr')

# Retrieve the top 30 entries (excluding the header row)
top_30_entries = rows[1:31]

# Initialize an empty list to store the extracted data
data = []

# Iterate over the rows and print the data
for row in top_30_entries:
    # Extract the columns for each row
    columns = row.find_elements(By.TAG_NAME, 'td')

    # Get the values from the columns
    Rank = columns[0].text.strip()
    Year = columns[1].text.strip()
    Title = columns[2].text.strip()
    Gross = columns[3].text.strip()
    
    # Append the data to the list
    data.append([Rank, Year, Title,Gross])

# Create a pandas DataFrame from the extracted data
Movies_original = pd.DataFrame(data, columns=['Rank', 'Year', 'Title', 'Gross'])

# Quit the browser
driver.quit()
```

HYPOTHESIS TESTING

```python
#Read dataframes from Excel
profitable_movies_books = pd.read_excel("profitable_movies_books.xlsx")
profitable_original_movies = pd.read_excel("profitable_original_movies.xlsx")
```

```python
import scipy.stats as st

#H0 average gross revenue between movies based on books and movies not based on books is equal
#H1 average gross revenue between movies based on books and movies not based on books is not equal

movies_not_based_on_books = profitable_original_movies['Gross']
movies_based_on_books = profitable_movies_books['Gross']

# Conduct the two-sample t-test
t_statistic, p_value = st.ttest_ind(movies_not_based_on_books, movies_based_on_books)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
'''
The p-value of 0.0172 suggests that this difference is statistically significant at a significance level of 0.05. 
Reject the null hypothesis and conclude that there is a significant difference in the average gross revenue between movies based on books and movies not based on books. 

'''
```

```python
'''
The t-statistic is positive, wich suggests that the the mean gross revenue of movies not based on books is higher than the mean gross revenue of movies based on books.

'''
```

```python
movies_not_based_on_books = profitable_original_movies['Gross']
movies_based_on_books = profitable_movies_books['Gross']

mean_not_based_on_books = movies_not_based_on_books.mean()
mean_based_on_books = movies_based_on_books.mean()

print("Mean gross revenue of movies not based on books:", mean_not_based_on_books)
print("Mean gross revenue of movies based on books:", mean_based_on_books)
```

```python
''' 
The mean gross revenue of movies not based on books is approximately $1,204,705,304.37, while the mean gross revenue of movies based on books is approximately $956,631,384.67.
The mean gross revenue of movies not based on books is higher than the mean gross revenue of movies based on books. 
This supports the conclusion that movies not based on books tend to have higher gross revenue.
'''
```

```python
#H0  the average ratings are equal between the two dataframes
#H1  the average ratings are not equal between the two dataframes

ratings_original_movies = profitable_original_movies['Rating']
ratings_movies_books = profitable_movies_books['Rating']

# Conduct the two-sample t-test
t_statistic, p_value = st.ttest_ind(ratings_original_movies, ratings_movies_books)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
''' 
The p-value is greater than the significance level of 0.05, indicating that we do not have sufficient evidence to reject the null hypothesis.
The average ratings of the movies in both dataframes are not significantly different. It implies that, in terms of ratings, there is no clear distinction between movies based on books and movies not based on books 
in this particular dataset.
'''
```

```python
# H0 the average number of votes is equal between the two dataframes
# H1 the average number of votes is not equal between the two dataframes

votes_original_movies = profitable_original_movies['Votes']
votes_movies_books = profitable_movies_books['Votes']

# Conduct the two-sample t-test
t_statistic, p_value = st.ttest_ind(votes_original_movies, votes_movies_books)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
''' 
In this case, since the p-value (0.2308) is greater than the significance level of 0.05, I do not reject the null hypothesis. This suggests that there is not a significant difference in the average number of votes 
between the profitable_original_movies and profitable_movies_books dataframes. It suggests that both sets of movies may have a similar level of popularity.
'''
```
